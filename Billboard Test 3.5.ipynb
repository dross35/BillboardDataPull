```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import datetime
```

```python
import datetime

def first_saturday(year):
    """
    Finds the date of the first Saturday of a given year.
    """
    d = datetime.date(year, 1, 1)
    days_to_saturday = (5 - d.weekday()) % 7
    return d + datetime.timedelta(days=days_to_saturday)
n=[]
# Example usage:
for year in range(1990, 2024):
    n.append(first_saturday(year))
```

```python
jj=0
while jj<len(n):
    n[jj]=str(n[(jj)])
    jj+=1
```
