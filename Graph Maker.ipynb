```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def bine(h,j,r):
    f=pd.read_excel(h.format(j))
    nsl=['(',"",')',"",'[',"",']',"","Tyler, The Creator","Tyler The Creator",'Kanye West Jay Z','Kanye West,Jay Z','Jay Z Kanye West','Jay Z,Kanye West','Pusha T Kanye West','Pusha T,Kanye West','Silk Sonic Bruno Mars','Bruno Mars','Jay Z','Jay-Z',',Jr','Jr','Migos','Quavo,Offset,Takeoff','Uncle Charlie Wilson','Charlie Wilson','JAY Z','Jay-Z','Mulatto','Latto','NAS','Nas','Lil Bow Wow','Bow Wow','Young Jeezy','Jeezy','DJ Mustard','Mustard','K Camp','Kcamp','IamSu!','IamSu','IamSU!','IamSu','Rae Sremmurd','Swae Lee,Slim Jxmmi','Pharrell Williams','Pharell','Kevin K-MAC McCall','Kevin McCall','Boosie BadAzz',"Lil' Boosie",'NBA YoungBoy','YoungBoy Never Broke Again','Youngboy Never Broke Again','YoungBoy Never Broke Again','Missy "Misdemeanor" Elliott','Missy Elliott', ', Jr.','Jr.','Ye','Kanye West','Wizkid','WizKid','Jay-Z','JAY-Z','Lipps, Inc.','Lipps Inc.',', Sr','Sr','A Boogie Wit da Hoodie','A Boogie Wit Da Hoodie','Bootsy','Bootsy Collins','William Bootsy Collins','Bootsy Collins','Daryl Hall John Oates','Daryl Hall,John Oates']
    asl=[' & ', ' Featuring ',', ',' Duet With ',' Introducing ',' With ',' X ',' x ',' + ',' or ',' Or ',' Feat. ',' And ',' featuring ',' Duet ',' Presents ',' feauting ',' Feauting ',' Starring ',' Featuing ',' Feat.',' Feautring ',' F/',' Co-Starring ',' with ','/',' and ',' Feturing ']
    for n in range(j+1,r+1):
        l=pd.read_excel(h.format(n))
        f=pd.concat([f,l])
    aa=f.sort_values(['Song','Artists'],ascending=True).loc[:,['Song','Rank','Artists','Pts','Date']].values.tolist()
    nnt=0
    nnj=[]
    while nnt<len(aa)-1:
        if (aa[nnt][0]==aa[nnt+1][0]) and (aa[nnt][2]==aa[nnt+1][2]):
            aa[nnt+1][3]+=aa[nnt][3]
        else:
            nnj.append(aa[nnt])
        nnt+=1
    nnj.append(aa[nnt])
    cnn=0
    ljj=[]
    while cnn<len(nnj):
        cd=str(nnj[cnn][2])
        h=0
        while h<len(nsl)-1:
            cd=cd.replace(nsl[h],nsl[h+1])
            h+=2
        h=0
        while h<len(asl):
            cd=cd.replace(asl[h],',')
            h+=1
        b=cd.split(',')
        nj=0
        if len(b)>1:
            while nj<len(b):
                opp=[nnj[cnn][0],b[nj],nnj[cnn][3]]
                ljj.append(opp)
                nj+=1
        else:
            opp=[nnj[cnn][0],nnj[cnn][2],nnj[cnn][3]]
            ljj.append(opp)
        cnn+=1
    ljj=pd.DataFrame(ljj,columns=['Song','Artists','Pts'])
    nk=ljj.sort_values('Artists',ascending=True).values.tolist()
    while nk[0][1]=='':
        if nk[0][1]=='':
            del nk[0]
    mlk=0
    gt=1
    total=[]
    while mlk<len(nk)-1:
        if nk[mlk][1]==nk[mlk+1][1]:
            nk[mlk+1][2]+=nk[mlk][2]
            gt+=1
        else:
            nk[mlk].append(gt)
            total.append(nk[mlk])
            if mlk==len(nk)-2:
                gt=gt
            else:
                gt=1
        mlk+=1
    nk[mlk].append(gt)
    total.append(nk[mlk])
    totals=[]
    mlk=0
    while mlk<len(total):
        jh=total[mlk][1:]
        totals.append(jh)
        mlk+=1
    f=pd.DataFrame(totals,columns=['Artist','Pts','Songs'])
    return f
```

```python
lk=0
while lk!=1 and lk!=2:
    lk=int(input('1 for single 2 for range:'))
    dh=str(input('choose a category country,100,rap,or hhrb:'))
if lk==1:
    nk=int(input("What year you want to look at:"))
    ll=int(input("How many artists:"))
    nm=int(input("At which rank do you want the artists to start at:"))
    ll+=nm
    if dh=='country':
        jj=int(input('10,25,or 50:'))
        if jj==10:
            kk='Country/t10/years/al/{}.xlsx'.format(nk)
        elif jj==25:
            kk='Country/t25/years/al/{}.xlsx'.format(nk)
        elif jj==50:
            kk='Country/t50/years/al/{}.xlsx'.format(nk)
    elif dh=='100':
        jj=int(input('10,25,50,or 100:'))
        if jj==10:
            kk='Hot100/t10/years/al/{}.xlsx'.format(nk)
        elif jj==25:
            kk='Hot100/t25/years/al/{}.xlsx'.format(nk)
        elif jj==50:
            kk='Hot100/t50/years/al/{}.xlsx'.format(nk)
        elif jj==100:
            kk='Hot100/t100/years/al/{}.xlsx'.format(nk)
    elif dh=='rap':
        jj=int(input('10 or 25:'))
        if jj==10:
            kk='Rap/t10/years/al/{}.xlsx'.format(nk)
        elif jj==25:
            kk='Rap/t25/years/al/{}.xlsx'.format(nk)
    elif dh=='hhrb':
        jj=int(input('10,25,or 50:'))
        if jj==10:
            kk='HHRB/t10/years/al/{}.xlsx'.format(nk)
        elif jj==25:
            kk='HHRB/t25/years/al/{}.xlsx'.format(nk)
        elif jj==50:
            kk='HHRB/t50/years/al/{}.xlsx'.format(nk)
    fd=pd.read_excel(kk).iloc[:,1:]
elif lk==2:
    nk=int(input("What year you want to start at:"))
    nk1=int(input("What year you want to end at:"))
    ll=int(input("How many artists:"))
    nm=int(input("At which rank do you want the artists to start at:"))
    ll+=nm
    if dh=='country':
        jj=int(input('10,25,or 50:'))
        if jj==10:
            kk='Country/t10/years/sn/{}song.xlsx'
        elif jj==25:
            kk='Country/t25/years/sn/{}song.xlsx'
        elif jj==50:
            kk='Country/t50/years/sn/{}song.xlsx'
    elif dh=='100':
        jj=int(input('10,25,50,or 100:'))
        if jj==10:
            kk='Hot100/t10/years/sn/{}song.xlsx'
        elif jj==25:
            kk='Hot100/t25/years/sn/{}song.xlsx'
        elif jj==50:
            kk='Hot100/t50/years/sn/{}song.xlsx'
        elif jj==100:
            kk='Hot100/t100/years/sn/{}song.xlsx'
    elif dh=='rap':
        jj=int(input('10 or 25:'))
        if jj==10:
            kk='Rap/t10/years/sn/{}song.xlsx'
        elif jj==25:
            kk='Rap/t25/years/sn/{}song.xlsx'
    elif dh=='hhrb':
        jj=int(input('10,25,or 50:'))
        if jj==10:
            kk='HHRB/t10/years/sn/{}song.xlsx'
        elif jj==25:
            kk='HHRB/t25/years/sn/{}song.xlsx'
        elif jj==50:
            kk='HHRB/t50/years/sn/{}song.xlsx'
    fd=bine(kk,nk,nk1)
if ll!=1:
    fda=fd.nlargest(ll,'Pts').iloc[nm-1:nm+ll-2]
else:
    fda=fd.iloc[:,:]
a,b,c=fda.iloc[:,0],fda.iloc[:,1],fda.iloc[:,2]
ca,ba=np.array(c),np.array(b)
m,d=np.polyfit(ca,ba,1)

```

```python
fig,jk=plt.subplots()
jk.scatter(c,b,color='g')
jk.plot(c,m*c+d,color='gray',linestyle=':')
jk.axhline(np.mean(ba),color='blue',linestyle='--')
jk.axvline(np.mean(ca),color='red',linestyle='--')
plt.xlabel('Songs')
plt.ylabel('Points')
if lk==1:
    plt.title('{} top {} artists'.format(nk,ll-1))
elif lk==2:
    plt.title('{}-{} Top {} Artists'.format(nk,nk1,ll-nm))
for i in (range(len(fda))):
    jk.annotate(a.iloc[i], (c.iloc[i], b.iloc[i]))
print(fda.sort_values(['Pts','Songs'],ascending=False))
```

```python
import numpy as np

def remove_outliers_IQR(data):
    Q1, Q3 = np.percentile(data, [25, 75])
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    return data[(data >= lower) & (data <= upper)]

data = np.array([1,2,3,4,5,6,100])
ba = remove_outliers_IQR(ba)
ba
```
